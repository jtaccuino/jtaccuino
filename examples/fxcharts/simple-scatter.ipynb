## Scatter Plot

```java
use(JTExtension.FX_CHARTS_EXTENSION)
```

```java
var r = new Random();
```

```java
var y = r.doubles(105, -5, 10).toArray();
var x = r.ints(105, 23, 98).toArray();
```

```java
var chart = scatterFx(x,y, "Random X", "Random Y", "Random data");
display(chart);
```

```java
record DataPoint(int x, double y) {}
```

```java
var dataPoints = IntStream.range(0, x.length).mapToObj(i -> new DataPoint(x[i],y[i])).collect(Collectors.groupingBy(d -> d.y() >0))
```

```java
var positiveValues = dataPoints.get(Boolean.TRUE);
var negativeValues = dataPoints.get(Boolean.FALSE);
```

```java
double[][] convertToDoubles(List<DataPoint> dataPoints) {
var values = new double[2][dataPoints.size()];
for (int i=0; i < dataPoints.size(); i++) {
  values[0][i] = dataPoints.get(i).x();
  values[1][i] = dataPoints.get(i).y();
}
return values;
}
```

```java
var posArray = convertToDoubles(positiveValues);
var negArray = convertToDoubles(negativeValues);
negArray[1]
```

```java
var multiChart = scatterPlot(List.of(posArray[0],negArray[0]),List.of(posArray[1],negArray[1]) ,
            "x Values", "y Values", new String[]{"pos","neg"});
display(multiChart)
```
